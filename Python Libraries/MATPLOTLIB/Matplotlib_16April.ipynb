```python
# a problem how to remove the overlapping of x axis

children = [10,20,40,30,20]
colors = ['red red red red red ','blue blue blue blue blue blue','green green green green','yellow yellow yellow', 'pink']
plt.bar(colors,children)

plt.xticks(rotation = 'vertical')
```

```python
# Stacked Bar Chart

plt.bar(df['batsman'],df['2017'], label = '2017')
plt.bar(df['batsman'],df['2016'],bottom = df['2017'], label = '2016')
plt.bar(df['batsman'],df['2015'], bottom = (df['2016'] + df['2017']),label='2015')

plt.legend()
plt.show()
```

# 5. Pie Chart

image.png

* Univariate /Bivariate Analysis

* Categorical VS Numerical

* Use case: to find the contribution on a standard scale

```python
import matplotlib.pyplot as plt
import pandas as pd
data = [10,23,45,23,89]
subject = ['Eng','Hindi','maths','Science','Social']

plt.pie(data,labels = subject)
plt.show()
```

```python
df = pd.read_csv('gaylerun.csv')

plt.pie(df['batsman_runs'],labels= df['batsman'],autopct='%.1f%%')

plt.show()
```

```python

plt.pie(df['batsman_runs'],labels= df['batsman'],autopct='%.1f%%',colors = ['blue','green','yellow','pink','cyan','red'])

plt.show()

plt.savefig('sample.png')
```

```python

# explode shadow

plt.pie(df['batsman_runs'],labels= df['batsman'],autopct='%.1f%%',explode = [0.1,0,0.3,0,0,0],shadow = True)

plt.show()
```
