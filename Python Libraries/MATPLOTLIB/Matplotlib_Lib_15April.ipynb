```python
# Form a pandas dataframe

batsman = pd.read_csv('Batsman.csv')
batsman
```

```python
# Both the batsman analysis in a single graph
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# both ROhit sharma and V kohli together
# ALso we want to label the graph
# is in a single graph
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title("Rohit Sharma Vs V Kohli Career Comparisions")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

```python
# We can chnage the colour of the 2D graph lines
plt.plot(batsman['index'],batsman['V Kohli'], color='Green')
plt.plot(batsman['index'],batsman['RG Sharma'], color='Black')

plt.title("Rohit Sharma Vs V Kohli Career Comparisions")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

```python
# we can change the line width and style
plt.plot(batsman['index'],batsman['V Kohli'],linestyle='dashed')
plt.plot(batsman['index'],batsman['RG Sharma'],linestyle='dotted')

plt.title("Rohit Sharma Vs V Kohli Career Comparisions")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

```python
# We can change the color line width and the line style
plt.plot(batsman['index'],batsman['V Kohli'],color ='red',linestyle='dashed',linewidth=2)
plt.plot(batsman['index'],batsman['RG Sharma'],color = 'green', linestyle= 'dotted',linewidth=3)

plt.title("Rohit Sharma Vs V Kohli Career Comparisions")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

# CHANGE THE COLOR , lINE WIDTH AND LINE STYLE,  Marker Size and marker

```python


plt.plot(batsman['index'],batsman['V Kohli'] , color ='red', linestyle='dashed',linewidth=2,marker = 'D', markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color = 'green', linestyle= 'dotted',linewidth=3,marker='o',markersize=5)

plt.title("Rohit Sharma Vs V Kohli Career Comparisions")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

```python
# Limiting axes

# plotting a simple function

price = [48000, 54000, 57000, 49000,47000, 450000]
year = [2015, 2016,2017,2018,2019,2020]

plt.plot(year, price)


plt.ylim(0,75000)
plt.xlim(2017,2019)
```

```python
# plotting a scatter diagram on a proper dataframe

df2 = pd.read_csv('batter.csv')
df2

df2=df2.head(50)
plt.scatter(df2['avg'], df2['strike_rate'])
plt.title("Average and Strike Rate of top 50 batsman")
```

```python
# plotting a scatter diagram on a proper dataframe



plt.scatter(df2['avg'], df2['strike_rate'] , color='red',marker='*')
plt.title("Average and Strike Rate of top 50 batsman")
```

```python
# size

tips = sns.load_dataset('tips')
tips

plt.scatter(tips['total_bill'],tips['tip'],s=tips['size']*20, marker='o',color='green')
plt.title("Average and strike rate of top 50 batsman ")
plt.xlabel("Average")
plt.ylabel("Strike Rate")
```
